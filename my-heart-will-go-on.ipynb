```r
ds <- read.csv("../input/titanic/train.csv")
ds
```

```r
oldest <-  which.max(ds$Age)# Age == max(Age))
#oldest
oldest_person <- ds[oldest,]$Name
youngest <-  which.min(ds$Age)# Age == max(Age))

youngest_person <- ds[youngest,]$Name
#oldest_person
#youngest_person

old_str <- "The oldest person is"
young_str <- "The youngest person is"

ager <- "at age"
closer <- "years"
paste(young_str,youngest_person ,ager, ds[youngest,]$Age,closer)
paste(old_str,oldest_person ,ager, ds[oldest,]$Age,closer)

#class(ds)
```

```r
#ds[ds$Sex  == "male"] #[Sex]

males <- subset(ds, Sex == "male")
male_survivors <- subset(males, Survived == 1)

male_survivors_1 <- subset(male_survivors, Pclass == 1)
male_survivors_2 <- subset(male_survivors, Pclass == 2)
male_survivors_3 <- subset(male_survivors, Pclass == 3)

male_survivors_1_len <- nrow(male_survivors_1)
male_survivors_2_len <- nrow(male_survivors_2)
male_survivors_3_len <- nrow(male_survivors_3)


main_len <- nrow(ds)
male_survivors_len <- nrow(male_survivors)
male_len <-nrow(males)
male_survivors_percent <- male_survivors_len/male_len

male_survivors_1_percent <- male_survivors_1_len/male_len
male_survivors_2_percent <- male_survivors_2_len/male_len
male_survivors_3_percent <- male_survivors_3_len/male_len

male_survivors_percent
male_survivors_1_percent
male_survivors_2_percent 
male_survivors_3_percent
#male_survivors
#males

```

```r
male_fares <- males$Fare #["Fare"]#subset(males, is.double(Fare))
male_fares_mean <- mean(male_fares) #male_fares_mean = mean(male_fares)
male_fares_mean
```

```r
library(plotrix)
slices <- c(male_survivors_1_len, male_survivors_3_len, male_survivors_2_len)
labels <- c("1st class", "3rd class", "2nd class")
pie3D(slices,labels=labels,explode=0.0,
   main="Pie Chart of male survivors in each class",theta=1.24)
```

```r
females <- subset(ds, Sex == "female")
female_survivors <- subset(females, Survived == 1)

female_survivors_1 <- subset(female_survivors, Pclass == 1)
female_survivors_2 <- subset(female_survivors, Pclass == 2)
female_survivors_3 <- subset(female_survivors, Pclass == 3)

female_survivors_1_len <- nrow(female_survivors_1)
female_survivors_2_len <- nrow(female_survivors_2)
female_survivors_3_len <- nrow(female_survivors_3)


main_len <- nrow(ds)
female_survivors_len <- nrow(female_survivors)
female_len <-nrow(males)
female_survivors_percent <- female_survivors_len/male_len

female_survivors_1_percent <- female_survivors_1_len/male_len
female_survivors_2_percent <- female_survivors_2_len/male_len
female_survivors_3_percent <- female_survivors_3_len/male_len

female_survivors_percent
female_survivors_1_percent
female_survivors_2_percent 
female_survivors_3_percent
```
